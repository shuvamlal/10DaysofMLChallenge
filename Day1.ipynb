```python
res = []
for t in l.columns[4:]:
    n = [1 for i in l[t] if i==0]
    if sum(n) >= (0.7*len(l[t])):
        res.append(t)
#     l.drop(t, axis=1)
```

```python
final = l.drop(res, axis=1)
```
